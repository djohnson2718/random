```python
import sage.libs.lrcalc.lrcalc as lrcalc
```

```python
def check(m, k=0):
    total = 0
    lam = 3*m-2*k, m + 2*k
    for alpha in Partitions(m,max_length=2):
        talpha = [2*ai for ai in alpha]
        tgam = copy(talpha)
        tgam[0] = tgam[0]+2
        
        for beta in Partitions(m,max_length=2):
            tbeta = [2*bi for bi in beta]            
            c=lrcalc.lrcoef(lam,talpha,tbeta)
            print "albe", talpha, tbeta, c
            total += c
            if len(tbeta)==2 and tbeta[0]==tbeta[1]:
                print "no corresponding 0"
            else:
                tdel = copy(tbeta)
                tdel[0]=tdel[0]-2
                c=lrcalc.lrcoef(lam,tgam,tdel)
                print "gade", tgam,tdel, c
                total -= c

            print
        print "***************"
        
    if is_odd(m):
        print "extra for odd m!"
        tgam = [m+1,m+1]
        for delta in Partitions(m-1,max_length=2):
            tdel = [2*di for di in delta]
            c=lrcalc.lrcoef(lam,tgam,tdel)
            print "gabe",tgam,tdel, c
            total -= c
    
        print "*************"
        
    if total == 1:
        print "Sum is 1 as expected!"
    else:
        print "WARNING, sum is {}".format(total)
```
