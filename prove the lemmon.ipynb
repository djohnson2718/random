```python
def do_skew(mu,nu):
    n = sum(mu)
    RR = PolynomialRing(QQ, ["x"+str(i) for i in range(n)])
    
    sst=SemistandardSkewTableaux([mu,nu], max_entry=n)

    w=RR.gens()
    p=0
    for t in sst:
        p += prod([v^e for v, e in zip(w,t.evaluation())])
    #print p
    #print s[2,2,1].expand(5)(*w)
    #print p - s[2,2,1].expand(5)(*w)
    print s.from_polynomial(p)
```
